```python
# Replace null values in 'genres' column with 'Horror, Thriller' where 'runtime_minutes' is < 100
df1.loc[df1['runtime_minutes'] < 100, 'genres'] = df1.loc[df1['runtime_minutes'] < 100, 'genres'].fillna('Horror, Thriller')
df1.loc[df1['runtime_minutes'] >= 100, 'genres'] = 'Documentary'
```

```python
# Considered introducing the directors table because directors play a key role in film production
df6 = pd.read_sql("""SELECT * FROM directors;""", conn)
df6.head()
```

```python
# Change the column names to uppercase
df6.columns = [column.upper() for column in df6.columns]
df6.head()
```

```python
# Merging df5 table with the df6 table to include directors 'id' 
```

```python
finaldf = df5.merge(df6, on='MOVIE_ID', how='inner')
finaldf
```

```python
# To remove duplicates from specific columns (MOVIE_ID, PRIMARY_TITLE, START_YEAR, RUNTIME_MINUTES, and PERSON_ID)
finaldf.drop_duplicates(subset=['MOVIE_ID', 'PRIMARY_TITLE', 'START_YEAR', 'RUNTIME_MINUTES', 'PERSON_ID'], inplace=True)
finaldf
```

```python
# remame the PERSON'S ID TO DIRECTOR'S_ID
finaldf.rename(columns={'PERSON_ID': 'DIRECTOR_ID'}, inplace=True)
finaldf.head()
```

```python
import matplotlib.pyplot as plt

# reducing the sample to 20 for clarity
director_ratings = finaldf.groupby('DIRECTOR_ID')['AVERAGERATING'].mean()
top_20_directors = director_ratings.sort_values(ascending=False).head(20)

# Plotting the top 20 directors and their average ratings
plt.figure(figsize=(12, 6))
plt.plot(top_20_directors.index, top_20_directors.values, marker='o', linestyle='-', color='b')

plt.title('Top 20 Directors by Average Movie Rating')
plt.xlabel('Director ID')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()

plt.show()
```
